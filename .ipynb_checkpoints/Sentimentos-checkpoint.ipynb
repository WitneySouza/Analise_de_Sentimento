```python
df = pd.read_csv('br_gov_meugovbr.csv')
df.head()
```

```python
#criando as variáveis de treinamento e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Review, df.NegativeRating, test_size=0.3)
print(X_train, X_test, y_train, y_test)
```

```python
#transformando texto em vetores
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:10]
```

```python
#treinando o modelo
from sklearn.naive_bayes import MultinomialNB #importando o modelo Naive Bayes Multinomial
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
#testanto a classificação de novos elementos
ratings = [
    'Isso torna as coisas mais fáceis',
    'Por favor, preciso de ajuda. Não consigo entrar no sistema, mensagem de erro "tente mais tarde" durante toda a semana. Não há suporte com quem eu fale. Se você é desenvolvedor, por favor me sugira, o que fazer?'
]
ratings_count = v.transform(ratings)
model.predict(ratings_count)
```

```python
#acurácia do modelo
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
#criar uma matriz de confusão
from sklearn.metrics import confusion_matrix, accuracy_score 
previsoes = model.predict(X_test_count)
print(confusion_matrix(y_test, previsoes))
```

```python
#ver métricas do modelo
from sklearn import metrics
print(accuracy_score(y_test, previsoes))
print(metrics.classification_report(y_test, previsoes))
```
